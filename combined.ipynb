# Data Aggregation

Loading Feature Importance Data of Dataset 1

```python
feature_importances1 = pd.read_csv("importances/feature_importances1.csv")
feature_importances1.index = feature_importances1['Feature']
```

Loading Feature Importance Data of Dataset 2


```python
feature_importances2 = pd.read_csv("importances/feature_importances2.csv")
feature_importances2.index = feature_importances2['Feature']
```

Loading Feature Importance Data of Dataset 3

```python
feature_importances3 = pd.read_csv("importances/feature_importances3.csv")
feature_importances3.index = feature_importances3['Feature']
```

Plotting Feature Importance Graph for Dataset 1

```python
fig, ax = plt.subplots()
feature_importances1.plot.bar( ax=ax)
ax.set_title("Feature importances Dataset 1")
ax.set_ylabel("Feature Importance in %")
fig.tight_layout()
```

Plotting Feature Importance Graph for Dataset 2

```python
fig, ax = plt.subplots()
feature_importances2.plot.bar( ax=ax)
ax.set_title("Feature importances Dataset 2")
ax.set_ylabel("Feature Importance in %")
fig.tight_layout()
```

Plotting Feature Importance Graph for Dataset 3

```python
fig, ax = plt.subplots()
feature_importances3.plot.bar(ax=ax)
ax.set_title("Feature importances Dataset 3")
ax.set_ylabel("Feature Importance in %")
fig.tight_layout()
```

```python
# Display feature_importances1 DataFrame

feature_importances1
```

```python
# Get the number of rows in feature_importances1 DataFrame
feature_importances1.shape[0]
```

```python
# Modify feature names and normalize importance values for Dataset 1
feature_importances1["Feature"] = feature_importances1["Feature"]+"_1"
feature_importances1["Normalised_Importance"] = feature_importances1["Importance"]/feature_importances1["Importance"].max()
feature_importances1
```

```python
# Modify feature names and normalize importance values for Dataset 2
feature_importances2["Feature"] = feature_importances2["Feature"]+"_2"
feature_importances2["Normalised_Importance"] = feature_importances2["Importance"]/feature_importances2["Importance"].max()
feature_importances2
```

```python
# Display modified feature_importances3 DataFrame

feature_importances3["Feature"] = feature_importances3["Feature"]+"_3"
feature_importances3["Normalised_Importance"] = feature_importances3["Importance"]/feature_importances3["Importance"].max()
feature_importances3
```

```python
# Combine feature importance data from all datasets

combined_feature_importance = pd.concat([feature_importances1, feature_importances2, feature_importances3], ignore_index=True)
combined_feature_importance = combined_feature_importance.drop(columns=['Importance',])
combined_feature_importance = combined_feature_importance.sort_index()
combined_feature_importance = combined_feature_importance.sort_values(by="Normalised_Importance")
combined_feature_importance.index = combined_feature_importance['Feature']
combined_feature_importance
```

```python
# Plot combined feature importance graph

fig, ax = plt.subplots()
combined_feature_importance.plot.bar(ax=ax)
ax.set_title("Feature importances")
ax.set_ylabel("Feature Importance in %")
fig.tight_layout()
fig.set_size_inches(10,8)
```
